```python
import numpy as np 
import matplotlib.pyplot as plt 
import time

%matplotlib inline

import pylab as pl 
from IPython import display

np.random.seed(1234)
```

```python
m = 100
x1 = 8 * np.random.rand(m, 1)
x2 = 7 * np.random.rand(m, 1) - 4 

g = 0.8 * x1 + x2 - 3
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x1, x2)
plt.show()
```

```python
C0 = np.where(g < -1)[0] # class 0 
C1 = np.where(g >= 1)[0] # class 1 
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x1[C0], x2[C0], 'bo', label='C0: -1')
plt.plot(x1[C1], x2[C1], 'ro', label='C1: +1')
plt.title('Linearly Separable Classes')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```

```python
M0 = C0.shape[0]
M1 = C1.shape[0]
```

```python
print("number of class 0 (-1) ->", M0)
print("number of class 1 ( 1) ->", M1)
```

```python
X0 = np.hstack([np.ones([M0, 1]), x1[C0], x2[C0]])
X1 = np.hstack([np.ones([M1, 1]), x1[C1], x2[C1]])

X = np.vstack([X0, X1])

y = np.vstack([-np.ones([M0, 1]), np.ones([M1, 1])])

X = np.asmatrix(X)
y = np.asmatrix(y)
```

```python
print("shape of X:", X.shape)
print("shape of y:", y.shape)
```

```python
from sklearn.utils import shuffle

X, y = shuffle(X, y)

print("shape of X:", X.shape)
print("shape of y:", y.shape)
```

```python
w = np.zeros([3, 1])   # random initialization
w = np.asmatrix(w)

data_points = M0 + M1

for k in range(3):
    for i in range(data_points):
        
        figure, ax = plt.subplots(figsize=(10, 6))
        
        if y[i, 0] != np.sign(X[i, :] * w)[0, 0]:
            w = w + y[i, 0] * X[i, :].T   # (3 X 1) == (1 X 1) * (3 X 1)
            
            # plot graph
            pl.scatter(X[i, 1], X[i, 2], alpha=0.5, c='purple', s=300)

            # show plot only when the decision boundary changes
            x1p = np.linspace(0, 8, 100).reshape(-1, 1)
            x2p = -w[1, 0] / w[2, 0] * x1p - w[0, 0] / w[2, 0]

            pl.plot(x1[C0], x2[C0], 'bo', label='C0: -1')
            pl.plot(x1[C1], x2[C1], 'ro', label='C1: +1')
            pl.plot(x1p, x2p, c='green', label='Perceptron linear decision boundary')
            pl.title(f'{i}-th data point in iteration {k}: Linearly Separable Classes')
            pl.xlabel('$x_1$')
            pl.ylabel('$x_2$')
            pl.xlim([-5, 15])
            pl.ylim([-10, 10])
            pl.legend(loc=2)

            display.clear_output(wait=True)
            display.display(pl.gcf())
            print("Found mis-classified datapoint!")
            time.sleep(0.001) 
            continue

        # plot graph
        pl.scatter(X[i, 1], X[i, 2], alpha=0.5, c='yellow', s=300)

        # show plot only when the decision boundary changes
        x1p = np.linspace(0, 8, 100).reshape(-1, 1)
        x2p = -w[1, 0] / w[2, 0] * x1p - w[0, 0] / w[2, 0]

        pl.plot(x1[C0], x2[C0], 'bo', label='C0: -1')
        pl.plot(x1[C1], x2[C1], 'ro', label='C1: +1')
        pl.plot(x1p, x2p, c='green', label='Perceptron linear decision boundary')
        pl.title(f'{i}-th data point in iteration {k}: Linearly Separable Classes')
        pl.xlabel('$x_1$')
        pl.ylabel('$x_2$')
        pl.xlim([-5, 15])
        pl.ylim([-10, 10])
        pl.legend(loc=2)

        display.clear_output(wait=True)
        display.display(pl.gcf())
        time.sleep(0.001) 
        
        pl.clf()
```
