```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = 100 
w = np.array([[-6], [2], [1]])
X = np.hstack([np.ones([m, 1]), 4 * np.random.rand(m, 1), 4 * np.random.rand(m, 1)])

w = np.asmatrix(w)
X = np.asmatrix(X)

y = 1 / (1 + np.exp(-X*w)) > 0.5   # sigmoid function 

C0 = np.where(y == True)[0]
C1 = np.where(y == False)[0]

y = np.empty([m,1])
y[C1] = 1
y[C0] = -1
y = np.asmatrix(y)
```

```python
plt.figure(figsize=(5, 3))
plt.plot(X[C0, 1], X[C0, 2], 'bo', label='C0')
plt.plot(X[C1, 1], X[C1, 2], 'ro', label='C1')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.xlim([-1, 5])
plt.ylim([-1, 5])
plt.show()
```

```python
# define sigmoid function 
# h == sigmoid function 

def h(x, w):
    return 1 / (1 + np.exp(-x*w))
```

```python
w = np.zeros([3, 1])
w = np.asmatrix(w)
```

```python
print(X.shape)
print(y.shape)
print(w.shape)
```

```python
lr = 1e-4

for _ in range(10000):
    for i in range(100):
        grad = y[i, 0] * X[i, :].T * (1 / (1 + np.exp(y[i, 0]*w.T*X[i, :].T))) 
        w = w + lr * grad
```

```python
print("w:\n", w)
```

```python
xp = np.linspace(0, 4, 100).reshape(-1, 1)
yp = -w[1, 0] / w[2, 0] * xp - w[0, 0] / w[2, 0]

plt.figure(figsize=(5, 3))
plt.plot(xp, yp, 'g', label='Decision boundary with Logistic Regression')
plt.plot(X[C0, 1], X[C0, 2], 'bo', label='C0')
plt.plot(X[C1, 1], X[C1, 2], 'ro', label='C1')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.xlim([-1, 5])
plt.ylim([-1, 5])
plt.title('Logistic Regression with Gradient Descent')
plt.show()
```
