```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HW4/RNN/train_pp.txt', header=None) 
valid_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HW4/RNN/valid_pp.txt', header=None) 
```

```python
train = np.array(train_df)
valid = np.array(valid_df)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(train[:, 0], train[:, 1], 'g-.', label='Training perplexity', alpha=0.7)
ax.plot(valid[:, 0], valid[:, 1], 'r.-', label='Validation perplexity', alpha=0.7)
ax.axhline(y = 164.74, color = 'b', linestyle = '-', label='Testing perplexity', alpha=0.7)
ax.set_title('Training, Validation, and Testing Perplexity')
ax.set_xlabel('Epochs')
ax.set_ylabel('Perplexity')
ax.set_xlabel('Epochs')
ax.grid()
ax.legend()
```
