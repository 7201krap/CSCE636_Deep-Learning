```python
import numpy as np
import matplotlib.pyplot as plt 

from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 200 
x1 = np.random.randn(n, 1)
x2 = np.random.randn(n, 1)
noise = 0.5 * np.random.randn(n, 1)

y = 2 + 1 * x1 + 3 * x2 + noise
```

```python
print("shape of x1", x1.shape)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.set_title('Data')
ax.set_xlabel('$X_1$')
ax.set_ylabel('$X_2$')
ax.set_zlabel('Y')
ax.scatter(x1, x2, y, label='Data')
# ax.view_init(30, 30)
plt.legend()
plt.show()
```

```python
X = np.hstack([x1**0, x1, x2])
X = np.asmatrix(X)

theta = (X.T * X).I * X.T * y
print("shape of theta:\n", theta)
```

```python
X1, X2 = np.meshgrid(np.arange(np.min(x1), np.max(x1), 0.5),
                     np.arange(np.min(x2), np.max(x2), 0.5))
YP = theta[0, 0] + theta[1, 0] * X1 + theta[2, 0] * X2
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.set_title('Multivariate Regression')
ax.set_xlabel('$X_1$')
ax.set_ylabel('$X_2$')
ax.set_zlabel('Y')

# original data
ax.scatter(x1, x2, y, label='Data')

# regression plane
ax.plot_wireframe(X1, X2, YP, color='green', alpha=0.5, \
                  label='Multivariate Regression Plane')

# ax.view_init(30, 30)
plt.legend()
plt.show()
```
