```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1, 1)
```

```python
# Let's visualise the data
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
M, N = x.shape
```

```python
X = np.hstack([np.ones([M, 1]), x])
X = np.hstack([x**0, x])
X = np.asmatrix(X)
```

```python
theta = (X.T * X).I * X.T * y
```

```python
print("theta", theta)
print("theta.shape", theta.shape)
```

```python
plt.figure(figsize=(10, 6))

# original plot 
plt.scatter(x, y, label='original data')

# new plot with a line 
xp = np.arange(0, 5, 0.01).reshape(-1, 1)
yp = theta[0, 0] + theta[1,0] * xp
plt.plot(xp, yp, 'r', label='regression')

# additional settings
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

### What happens when some outliers exist?

```python
x = np.vstack(  [ x, np.array([0.5,3.8]).reshape(-1, 1) ]  )
y = np.vstack(  [ y, np.array([3.9,0.3]).reshape(-1, 1) ]  )
```

```python
M, N = x.shape

X = np.hstack([x**0, x])
X = np.asmatrix(X)

theta = (X.T * X).I * X.T * y
```

```python
plt.figure(figsize=(10, 6))

# original plot 
plt.scatter(x, y, label='data with outliers')

# new plot with a line 
xp = np.arange(0, 5, 0.01).reshape(-1, 1)
yp = theta[0, 0] + theta[1,0] * xp
plt.plot(xp, yp, 'r', label='regression')

# additional settings
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

### Let's use sklearn!

```python
from sklearn import linear_model
```

```python
x = np.array([0.1, 0.4, 0.7, 1.2, 1.3, 1.7, 2.2, 2.8, 3.0, 4.0, 4.3, 4.4, 4.9]).reshape(-1, 1)
y = np.array([0.5, 0.9, 1.1, 1.5, 1.5, 2.0, 2.2, 2.8, 2.7, 3.0, 3.5, 3.7, 3.9]).reshape(-1, 1)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
# theta_1
reg.coef_ 
```

```python
# theta_0
reg.intercept_ 
```

```python
plt.figure(figsize=(10, 6))

# original plot 
plt.scatter(x, y, label='original data')

# new plot with a line 
xp = np.arange(0, 5, 0.01).reshape(-1, 1)
plt.plot(xp, reg.predict(xp), 'r', label='regression')

# additional settings
plt.title('Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
