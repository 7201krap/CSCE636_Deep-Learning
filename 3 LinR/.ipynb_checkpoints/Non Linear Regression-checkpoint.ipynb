```python
# Non-linear regression is the same as Linear regression for Non-linear data
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
n = 100 
x = -5 + 15 * np.random.rand(n, 1)
noise = 10 * np.random.randn(n, 1)

y = 10 + 1 * x + 2 * x**2 + noise 
```

```python
plt.figure(figsize = (10, 8))
plt.title("True x and y")
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x, y, label='data')
plt.legend()
plt.show()
```

```python
X = np.hstack([x**0, x, x**2])
X = np.asmatrix(X)

theta = (X.T * X).I * X.T * y
print("theta:\n", theta)
```

```python
xp = np.linspace(np.min(x), np.max(x))
yp = theta[0, 0] + theta[1, 0] * xp + theta[2, 0] * xp**2
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x, y, label='data')
plt.plot(xp, yp, label='Non Linear Regression', color='red')

plt.title('Non Linear Regression')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.grid()
plt.legend()
plt.show()
```
